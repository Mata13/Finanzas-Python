# Analizando indicadores

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker_symbol = "NVDA"
```

```python
# Descargar datos de NVDA
nvidia_data = yf.download(ticker_symbol, start="2020-01-01", end="2023-10-04")
```

```python
print(nvidia_data)
```

```python
print(nvidia_data.head(5))
```

```python
# Datos de cierre

plt.figure(figsize=(12, 6))
plt.plot(nvidia_data.index, nvidia_data['Close'], label=f'{ticker_symbol} Closing Price', color='deepskyblue') # index el la fecha
plt.title(f'{ticker_symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
